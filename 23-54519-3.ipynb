```python
# table driven agent
table = {
'Dark': 'Turn_ON',
'Bright': 'Turn_OFF'
}

def table_driven_agent(percept):
 print("Perception Received:", percept)
 action = table[percept]
 return action
import random

conditions = ['Dark', 'Bright']

while True:
 percept = random.choice(conditions)
 action = table_driven_agent(percept)
 print("Action Performed:", action)

 cmd = input("Continue? (yes/no): ")
 if cmd != "yes":
  break
```

```python
# Simple Reflex Agent
def simple_reflex_agent(percept):
    print("Perception Received:", percept)
    if percept == 'Dark':
        action = 'Turn_ON'
    else:
        action = 'Turn_OFF'
    return action


import random

conditions = ['Dark', 'Bright']

while True:
    percept = random.choice(conditions)
    action = simple_reflex_agent(percept)
    print("Action Performed:", action)

    cmd = input("Continue? (yes/no): ")
    if cmd != "yes":
        break
```
